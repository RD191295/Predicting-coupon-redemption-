```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv("train.csv")
pd.set_option('display.max_columns', None)
train_data.head()
```

```python
train_new_data = train_data.drop(["userID","couponID1","brand1","productGroup1","categoryIDs1","couponID2","brand2","productGroup2","categoryIDs2","couponID3","productGroup3","brand3","orderID","categoryIDs3","coupon1Used","coupon2Used","coupon3Used"],axis  = 1)
train_new_data.head()
```

```python
train_new_data.info()
```

```python
train_new_data["orderTime"] = train_new_data["orderTime"].astype("datetime64[ns]")
train_new_data["couponsReceived"] = train_new_data["couponsReceived"].astype("datetime64[ns]")
```

```python
train_new_data.info()
```

```python
train_new_data["orderTime_day"] = train_new_data.orderTime.dt.day
train_new_data["orderTime_month"] = train_new_data.orderTime.dt.month
train_new_data["orderTime_year"] = train_new_data.orderTime.dt.year

train_new_data["orderTime_hour"] = train_new_data.orderTime.dt.hour
train_new_data["orderTime_min"] = train_new_data.orderTime.dt.minute
train_new_data["orderTime_sec"] = train_new_data.orderTime.dt.second
```

```python
train_new_data["couponsReceived_day"] = train_new_data.couponsReceived.dt.day
train_new_data["couponsReceived_month"] = train_new_data.couponsReceived.dt.month
train_new_data["couponsReceived_year"] = train_new_data.couponsReceived.dt.year

train_new_data["couponsReceived_hour"] = train_new_data.couponsReceived.dt.hour
train_new_data["couponsReceived_min"] = train_new_data.couponsReceived.dt.minute
train_new_data["couponsReceived_sec"] = train_new_data.couponsReceived.dt.second
```

```python
train_new_data.head()
```

```python
# droping Order time and coupons recieved column
train_new_data = train_new_data.drop(["orderTime","couponsReceived"],axis =1)
train_new_data
```

```python
# checking data type
train_new_data.info()
```

```python
# checking null values 
train_new_data.isnull().sum()
```

```python
# defining X and Y
X = train_new_data.drop("CouponCombination", axis = 1)
y = train_new_data["CouponCombination"]
```

```python
# standard scaliing
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_Scaled = scaler.fit_transform(X)
```

 ## Logistic Regression 

```python
# Logistic Regression 
from sklearn.linear_model import LogisticRegression

classification_log_model = LogisticRegression()
```

```python
classification_log_model.fit(X_Scaled,y)
```

```python
y_pred = classification_log_model.predict(X_Scaled)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

# lightbgm model

```python
import lightgbm as lgb
train_data = lgb.Dataset(X, label=y)
```

```python
# Set hyperparameters
params = {
    'boosting_type': 'gbdt',
    'objective': 'multiclass',
    'num_class': 8,
    'metric': 'multi_logloss',
    'num_leaves': 31,
    'learning_rate': 0.05
}

# Train the model
num_rounds = 100
model = lgb.train(params, train_data, num_rounds)
```

```python
y_pred = model.predict(test_new_data)
```
